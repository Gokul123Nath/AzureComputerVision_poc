```python
import os
from dotenv import load_dotenv
from azure.cognitiveservices.vision.computervision import ComputerVisionClient
from azure.cognitiveservices.vision.computervision.models import VisualFeatureTypes
from msrest.authentication import CognitiveServicesCredentials

# Load environment variables from a .env file
load_dotenv()

# Azure Cognitive Services configuration
vision_endpoint = "https://imagereterval.cognitiveservices.azure.com/"
vision_key = "133ae38a5d1a4493aa327e0aa57eb76c"

# Set up Computer Vision client
vision_client = ComputerVisionClient(vision_endpoint, CognitiveServicesCredentials(vision_key))

def extract_image_features(image_path):
    with open(image_path, "rb") as image_file:
        features = vision_client.analyze_image_in_stream(image_file, visual_features=[VisualFeatureTypes.tags])
        return [tag.name for tag in features.tags]

def main():
    # Replace with the path to your query image
    query_image_path = "sample_cog.jpg"

    # Extract features from the query image
    image_features = extract_image_features(query_image_path)

    # Print the results
    print("Query Image Features:", image_features)

if __name__ == "__main__":
    main()
```

```python
import os
from dotenv import load_dotenv
from azure.cognitiveservices.vision.computervision import ComputerVisionClient
from msrest.authentication import CognitiveServicesCredentials

# Load environment variables from a .env file
load_dotenv()

# Azure Cognitive Services configuration
vision_endpoint = "https://imagereterval.cognitiveservices.azure.com/"
vision_key = "133ae38a5d1a4493aa327e0aa57eb76c"

# Set up Computer Vision client
vision_client = ComputerVisionClient(vision_endpoint, CognitiveServicesCredentials(vision_key))

def get_image_embedding(image_path):
    with open(image_path, "rb") as image_file:
        image_features = vision_client.describe_image_in_stream(image_file)
        return image_features

def main():
    # Replace with the path to your query image
    query_image_path = "sample_cog.jpg"

    # Get image embedding using Computer Vision API
    image_embedding = get_image_embedding(query_image_path)

    # Print the results
    print("Image Embedding:")
    print(image_embedding)

if __name__ == "__main__":
    main()
```

```python
import requests
import json

# Replace with your Azure AI Vision endpoint and key
endpoint = "https://imagereterval.cognitiveservices.azure.com/"
key = "133ae38a5d1a4493aa327e0aa57eb76c"

# URL of the image you want to use
image_url = "https://learn.microsoft.com/azure/ai-services/computer-vision/media/quickstarts/presentation.png"

# Construct the API URL
api_url = f"{endpoint}/computervision/retrieval:vectorizeImage?api-version=2023-02-01-preview&modelVersion=latest"

# Set up the headers
headers = {
    "Content-Type": "application/json",
    "Ocp-Apim-Subscription-Key": key
}

# Set up the request body
body = {
    "url": image_url
}

# Send the POST request
response = requests.post(api_url, headers=headers, data=json.dumps(body))

# Parse the response
data = response.json()

# The 'vector' field in the response contains the image's vector

print(data['vector'])
print(data)
```
